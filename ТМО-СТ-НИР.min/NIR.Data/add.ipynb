```python
# IMPORTS

import pandas as pd

from tabulate import tabulate

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def Load(name):
    return pd.read_csv(f'../ASA-Data/{name}.csv')
```

```python
def Save(dtf, name):
    output_file_path = f'../ASA-Data/{name}.csv'
    dtf.to_csv(output_file_path, index=False)
```

```python
def PrintColsFulness(dtf):
    table_data = []
    for column in dtf.columns:
        dtype = dtf[column].dtype
        percent_missing = dtf[column].isnull().mean() * 100
        table_data.append([column, dtype, f"{percent_missing:.4f}%"])

    print(tabulate(table_data, headers=['Column', 'dtype', 'Percent Missing'], tablefmt='pretty'))
```

```python
df = Load('dataset')
```

```python
df.head()
```

```python
PrintColsFulness(df)
```

```python
df = df.dropna(axis=1, how='all')
```

```python
lo = [ 'eth.src', 'eth.dst', 'ip.src', 'ip.dst' ]
df.drop(lo, axis=1, inplace=True)
```

```python
df.head()
```

```python
for column in df.columns:
    # Проверка типа данных столбца
    if df[column].dtype.kind in 'ifc':
        # Если тип данных числовой, заполнение пропущенных значений нулями
        df[column].fillna(0, inplace=True)
    else:
        # Если тип данных не числовой, заполнение пропущенных значений строкой '0'
        df[column].fillna('0', inplace=True)
```

```python
PrintColsFulness(df)
```

```python
df.head()
```

```python
for column in df.columns:
    if df[column].dtype.kind not in 'ifc':
        df[column], _ = pd.factorize(df[column])
```

```python
PrintColsFulness(df)
```

```python
for column in df.columns:
    mn = df[column].min()
    mx = df[column].max()
    sc = mx - mn
    if sc > 0:
        df[column] = (df[column] - mn) / (sc)
```

```python
df.head()
```

```python
Save(df, 'low_served')
```
